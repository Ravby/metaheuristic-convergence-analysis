# Rating interval bands for CEC 2022 problems

#### "Line Magics"

This (`%use lets-plot`) "line magic" will apply **Lets-Plot library** descriptor, which adds to your notebook all the boilerplate code necessary to create plots. `%useLatestDescriptors` is needed for function `ggtb()`, which enables zoom/pan interactivity on the plot.

```kotlin
%useLatestDescriptors
%use lets-plot
```

#### Results directory path

Define the path to the results directory for CEC 2022 experiments.

```kotlin
val resultsDirectory = "${System.getProperty("user.dir")}../../../../../../../experimental_results/CEC2022"
```

#### Load results

Load algorithm performance results from the specified directory.

```kotlin
val algorithms = si.um.feri.lpm.loadAlgorithmResults(si.um.feri.lpm.BenchmarkId.CEC2022, resultsDirectory)
```

#### Plot generation

Generate a plot for the loaded algorithms with specified x-axis breaks.

```kotlin
var plot = si.um.feri.lpm.generatePlot(
    algorithms = algorithms,
    width = 1500, height = 750,
    minX = 0.0, maxX = si.um.feri.lpm.BenchmarkManager.get(si.um.feri.lpm.BenchmarkId.CEC2022).k.toDouble() - 1,
    showVerticalLines = false
)
```

#### Display the plot

Add zoom and pan interactivity to the plot and display the plot.

```kotlin
plot += ggtb()
plot.show()
```

#### Save the plot

Save the plot as an SVG file in the specified directory.

```kotlin
ggsave(plot, path = resultsDirectory, filename = "CEC2022_rating_interval_bands.svg")
```
